```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data = pd.read_csv('/content/Ventas_totales.csv')
data.head(5)
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#todos son float por lo que se van a tratar como valores numéricos
data.info()
```

```python
data1 = data.copy()
```

```python
#Se sustituyen los valores nulos de la columna salon_ventas y la de
#subtotal_ventas_alimentos_bebidas usando la mediana.
#Esto con la finalidad de evitar algún tipo de sesgo por valores atípicos.

data1["salon_ventas"]=data1["salon_ventas"].fillna(
    round(data["salon_ventas"].median(),1)
    )
data1["subtotal_ventas_alimentos_bebidas"]=data1["subtotal_ventas_alimentos_bebidas"].fillna(
    round(data["subtotal_ventas_alimentos_bebidas"].median(),1)
    )
```

```python
#Se verifica que efectivamente no hay valores nulos en la columna salon_ventas
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Se sustituyen los valores de bebidas, panaderia y lácteos usando la media
#Se escoge la media debido a que es la medida más robusta en estos casos.
#Dado que son elementos con alto volumen de ventas, un solo outlier no va a
#afectar de manera significativa los resultados.
data1["bebidas"]=data1["bebidas"].fillna(
    round(data["bebidas"].mean(),1)
    )
data1["panaderia"]=data1["panaderia"].fillna(
    round(data["panaderia"].mean(),1)
    )
data1["lacteos"]=data1["lacteos"].fillna(
    round(data["lacteos"].mean(),1)
    )
```

```python
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Para las tarjetas se usó el método de rellenar con un string en concreto.
#Esto con la intención de conocer las tarjetas que fueron imputadas y evitar
#de esta manera confusiones entre si la tarjeta fue imputada o no. Así como si la
#misma es de crédito o de débito.
data1["tarjetas_debito"] =data1["tarjetas_debito"].fillna("8888 8888 8888")
data1["tarjetas_credito"] =data1["tarjetas_credito"].fillna("0000 0000 0000")
```

```python
#Para la categoría de otros opté por rellenar con el valor de 0 dado que no se
#a que se refiera y no quiero que sea un valor que pueda influir
#de ninguna manera.

data1["otros"]= data["otros"].fillna(0)
data1["otros_medios"]= data["otros_medios"].fillna(0)

```

```python
#A las columnas de almacen, carnes y verdulerías se les aplicó un ffill
#que es un llenado del ultimo valor no nulo hacia adelante
data1["almacen"] =data1["almacen"].fillna(method="ffill")
data1["carnes"] =data1["carnes"].fillna(method="ffill")
data1["verduleria_fruteria"] =data1["verduleria_fruteria"].fillna(method="ffill")
```

```python
#A las columnas de alimentos_preparados_rotiseria, indumentario_calzados_textiles_hogar
# y a electronicos_articulos_hogar, se les aplicó un back fill que significa que el valor nulo se imputa por el siguiente valor no nulo
data1["alimentos_preparados_rotiseria"] =data1["alimentos_preparados_rotiseria"].fillna(method="bfill")
data1["indumentaria_calzado_textiles_hogar"] =data1["indumentaria_calzado_textiles_hogar"].fillna(method="bfill")
data1["electronicos_articulos_hogar"] =data1["electronicos_articulos_hogar"].fillna(method="bfill")
```

```python
#Podemos corroborar que no queda ningún valor nulo dentro del código
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Se convierte el df data1 en un archivo .csv
data1.to_csv('VentasTotalesLimpio')
```
